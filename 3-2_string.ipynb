## 7. 문자열 포매팅

```python
# 1) 숫자 바로 대입
food = "I eat %d apples" %3
food
```

```python
# 2) 문자열 바로 대입
food = "I eat %s apples" %"four"
food
```

```python
# 3) 숫자 값을 나타내는 변수로 대입
number = 2
food = "I eat %d apples" %number
food
```

```python
# 4) 2개 이상의 값 넣기
num =10
day = 'three'

"I ate %d apples. so I was sick for %s days." %(num, day)
```

```python
# %s는 숫자도 문자열로 대입
"I have %s apples." %3
```

```python
"I have %d apples." %3.14
```

```python
"I have %f apples." %3.14
```

```python
"I have %s apples." %3.14
```

```python
# 포맷코드와 문자 %를 같이 사용하려면 %%
'Error is %d%%.' %98
```

## 10. format()함수를 사용한 포매팅

```python
# 숫자 바로 대입
'I eat {} apples.'.format(4)
```

```python
# 문자 바로 대입
'I eat {} apples.'.format('four')
```

```python
# 변수로 대입
num = 3
'I eat {} apples.'.format(num)
```

```python
'I eat {0} apples.'.format(4)
```

```python
# 2개 이상 값 넣기
num = 10
day = 'five'
"I ate {} apples. so I was sick for {} days.".format(num, day)
```

```python
"I ate {0} apples. so I was sick for {1} days.".format(num, day)
```

```python
"I ate {1} apples. so I was sick for {0} days.".format(num, day)
```

```python
# 이름으로 넣기
"I ate {number} apples. so I was sick for {day} days.".format(day=3, number=10)
```

```python
"I ate {} apples. so I was sick for {day} days.".format(10,day=3)
```

```python
# 왼쪽 정렬
'{}'.format('hi')
```

```python
'{:<10}'.format('hi')
```

```python
'{:>10}'.format('hi')
```

```python
'{:^10}'.format('hi')
```

```python
# 공백 채우기
'{:!^10}'.format('hi')
```

```python
'{:^10}'.format('hi')
```

```python
# 소수점 표현
'{:0.3f}'.format(3.156458)
```

```python
'{:10.2f}'.format(3.145456)
```

## 11. f문자열 포매팅

```python
# 문자열 앞에 f접두사를 붙이기
name = '강민식'
age = 31
f'나의 이름은 {name}입니다. 나이는 {age}입니다.'
```

```python
# 표현식 사용하기
f'나는 내년이면 {age +1}살이 됩니다.'
```

```python
d = {'name':'강민식2', 'age' : 33}
f'나의 이름은 {d['name']}입니다. 나이는 {d['age']} 입니다.'
```

```python
# 정렬
f'{'hi':>10}'
```

```python
f'{'hi':<10}'
```

```python
f'{'hi':^10}'
```

```python
f'{'hi':#^10}'
```

```python
# 소수점 표현하기
f'{5.868526:0.2f}'
```

```python
f'{5.868526:10.2f}'
```

```python
f'{5.868526:<10.2f}'
```

```python
f'{5.868526:-10.2f}'
```

```python
f'{5.868526:#<10.2f}'
```
