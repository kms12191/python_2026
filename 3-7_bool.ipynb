## 2. 자료형의 참과 거짓이 프로그램에서 어떻게 쓰일까?

```python
a = [1,2,3,4]

while a:
    print(a.pop())
```

```python
a
```

```python
bool(a)
```

```python
# 조건문(if)
if []:
    print('참')
else:
    print('거짓')
```

```python
if [1,2,3]:
    print('참')
else:
    print('거짓')
```

```python
if [1]:
    print('참')
else:
    print('거짓')
```
