## 2. 리스트를 복사하고자 할 때

```python
a = [1,2,3]
b = a
b
```

```python
id(a)
```

```python
id(b)
```

```python
a is b #변수a, 변수b는 같은 객체를 가르킴
```

```python
# a에 있는 값을 변경
a[0] = 4
a
```

```python
b #a[0]변경을 했는데 b[0]도 변경된다.
```

### b변수에 a를 복사할 때 값만 복사하고 다른 주소를 가르키도록 만드는 방법
 - [:] 이용하기

```python
a = [1,2,3]
b = a[:]
b
```

```python
id(a)
```

```python
id(b)
```

```python
a[-1] = -3
a
```

```python
b
```

 - copy 모듈 이용하기

```python
from copy import copy
a = [1,2,3]
b = copy(a)
```

```python
b is a
```

```python
a[1] = 0
a
```

```python
b
```

 - copy() 함수 사용하기

```python
a = [1,2,3]
b=a.copy()
```

```python
id(a)
```

```python
id(b)
```

```python
a is b
```

```python
a[0] =-1
a
```

```python
b
```

## 3. 변수를 만드는 여러가지 방법

```python
# 튜플로 변수 만들기
a,b = ('python','Life')
```

```python
a
```

```python
b
```

```python
(a,b) = 'Python','Life'
```

```python
print(a,b)
```

```python
a,b = 'Python', 'Life'
```

```python
print(a,b)
```

```python
# 리스트로 변수 만들기
[a,b] = ['python','life']
a
```

```python
b
```

```python
# 여러 변수에 같은 값을 대입
a=b='Python'
```

```python
print(a,b)
```

```python
id(a)
```

```python
id(b)
```

```python
# 두 변수의 값 바꾸기(파이썬만 있는 기능)
a=3
b=5
```
